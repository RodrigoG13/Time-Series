```python
import matplotlib.pyplot as plt

# Datos de ejemplo de las posiciones del caminante aleatorio
positions = [(x, y, z) for x in range(3) for y in range(3) for z in range(3)]
print(positions)

# Total de celdas en la cmara (49x49x49)
total_cells = 2**3

# Conjunto para almacenar posiciones nicas
unique_positions = set()

# Lista para guardar el porcentaje de rea ocupada en cada paso del tiempo
occupied_percentage = []

# Calcular los porcentajes de ocupacin
for position in positions:
    unique_positions.add(position)
    # Calcular el porcentaje ocupado y agregarlo a la lista
    occupied_percentage.append(len(unique_positions) / total_cells * 100)

# Crear la grfica del tiempo vs porcentaje de rea ocupada
time_steps = list(range(1, len(occupied_percentage) + 1))
plt.figure(figsize=(10, 5))
plt.plot(time_steps, occupied_percentage, marker='o')
plt.title('Porcentaje de rea Ocupada por el Caminante Aleatorio en el Tiempo')
plt.xlabel('Paso de Tiempo')
plt.ylabel('Porcentaje de rea Ocupada (%)')
plt.grid(True)
plt.show()
```
